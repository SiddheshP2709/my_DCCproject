```python
import tabula
import pandas as pd
import csv
```

```python
pdf_path = "elec1.pdf"
tables = tabula.read_pdf(pdf_path, pages='all', multiple_tables=True)

concatenated_df = pd.concat(tables, ignore_index=True)
concatenated_df = concatenated_df.astype(str)
concatenated_df.to_csv("electoral1.csv", index=False)
print("Combined table saved as CSV.")
```

```python
electoral1 = pd.read_csv('electoral1.csv')
display(electoral1)
```
